## 2.Line Gragh - Gráfico de Linha

```python
# 1- Grafico de Linha Simples
# Supondo que nós queremos investigar a Receita (Revenue) pelo Tempo (Data), mas mudamos para Semanas_ID

ax = sns.lineplot(x='Week_ID', y='Revenue', data=raw_data) 
```

```python
# 2- Grafico de Linha Simples - Vamos colocar uma categoria para selecionar os dados
# Supondo que nós queremos investigar a Receita (Revenue) pelo Tempo (Data), mas mudamos para Semanas_ID
# Vamos adicionar o hue = legendas coloridas

ax = sns.lineplot(x='Week_ID', y='Revenue', data=raw_data, hue ='Promo') 
```

```python
#3- Grafico de Linha Simples - Vamos colocar uma categoria para selecionar os dados
# Supondo que nós queremos investigar a Receita (Revenue) pelo Tempo (Data), mas mudamos para Semanas_ID
# Vamos adicionar o hue = legendas coloridas
# Vamos alterar os estilos das linhas

ax = sns.lineplot(x='Week_ID', y='Revenue', data=raw_data, hue ='Promo', style='Promo') 
```

```python
#4- Vamos aumentar o tamnho do gráfico para faciliara a visualização e melhor entendimento e alterar os intervalos do tempo
sns.set(rc={'figure.figsize':(12,10)})
#ci=False (Deixo o gráfico sem sombra)
ax = sns.lineplot(x='Week_ID', y='Revenue', data=raw_data, hue ='Promo', style='Promo', ci=False) 
```

```python
#5- Vamos destacar os picos e vales do gráfico para melhor visualização
ax = sns.lineplot(x='Week_ID', y='Revenue', data=raw_data, hue ='Promo', style='Promo', ci=False, markers=True) 
```

```python
#6-Vamos agora alterar o tempo para os meses (mesmo gráfico acima)
ax = sns.lineplot(x='Month_ID', y='Revenue', data=raw_data, hue ='Promo', style='Promo', ci=False, markers=True) 
```

```python
#7-Vamos agora alterar o hue para os dias da semana (mesmo gráfico acima)
# Vamos adicionar o hue = legendas coloridas para os dias da semana
ax = sns.lineplot(x='Month_ID', y='Revenue', data=raw_data, hue ='Day_Name', style='Promo', ci=False, markers=True) 
```

```python
#8-Vamos agora alterar o hue para os dias da semana e o periodo de tempo para ano (mesmo gráfico acima)
# Vamos adicionar o hue = legendas coloridas para os dias da semana
ax = sns.lineplot(x='Year', y='Revenue', data=raw_data, hue ='Day_Name', style='Promo', ci=False, markers=True) 
```

## 3.Bar Plots - Gráfico de Barras

```python
#1- Gráfico de Barras: Receita Total (Total Revenue) x Mês (Month)
ax = sns.barplot(x='Month_ID', y='Revenue', data=raw_data)

# Vamos criar uma tabela para entendermos melhor o gráfico (Valores médios)
raw_data[['Revenue','Month_ID']].groupby('Month_ID').agg({'Revenue':'mean'})
```

```python
#2- Gráfico de Barras: Receita Total (Total Revenue) x Mês (Month) - Vamos retirar o  intervalo de confiança - Traço em cima das barras
# ci = confidence interval (intervalo de confiança)- Valores médios
ax = sns.barplot(x='Month_ID', y='Revenue', data=raw_data, ci=None)
```

```python
#3- Gráfico de Barras: Receita Total (Total Revenue) x Mês (Month) - Vamos retirar o  intervalo de confiança - Traço em cima das barras
# ci = confidence interval (intervalo de confiança)- Valores médios
# Vamos adicionar o hue = legendas coloridas para as Promoções
ax = sns.barplot(x='Month_ID', y='Revenue', data=raw_data, ci=None, hue='Promo')
```

```python
#4- Gráfico de Barras Horizontais: Receita Total (Total Revenue) x Mês (Month) - Vamos retirar o  intervalo de confiança - Traço em cima das barras
# Precisamos inverter os eixos e a orientação para horizontal
# ci = confidence interval (intervalo de confiança)- Valores médios
# Vamos adicionar o hue = legendas coloridas para as Promoções
ax = sns.barplot(x='Revenue', y='Month_ID', data=raw_data, ci=None, hue='Promo', orient='h')
```

```python
#5- Gráfico de Barras Horizontais: Receita Total (Total Revenue) x Ano (Year)
# Precisamos inverter os eixos e a orientação para horizontal
# ci = confidence interval (intervalo de confiança)- Valores médios
# Vamos adicionar o hue = legendas coloridas para as Promoções
ax = sns.barplot(x='Revenue', y='Year', data=raw_data, ci=None, hue='Promo', orient='h')
```

```python
#6- Gráfico de Barras Horizontais: Receita Total (Total Revenue) x Ano (Year) - Vamos mudar as cores das barras
# Precisamos inverter os eixos e a orientação para horizontal
# ci = confidence interval (intervalo de confiança)- Valores médios
# Vamos adicionar o hue = legendas coloridas para as Promoções
ax = sns.barplot(x='Revenue', y='Year', data=raw_data, ci=None, hue='Promo', orient='h', color = '#33F6FF')

#Podemos escolher as cores conforme o site:
#https://htmlcolorcodes.com/
# Selecionammos o valor HEX do site
```

```python
#1- Vamos plotar a distribuição das Receitas (Renevue) - Transformando em Array ou Lista
x = raw_data['Revenue'].values
sns.distplot(x, color='blue')
```

```python
#2- Vamos plotar a distribuição das Receitas (Renevue) dos Valores Médios- Transformando em Array ou Lista
media = raw_data['Revenue'].mean()
media
```

```python
#2- Vamos plotar a distribuição das Receitas (Renevue) dos Valores Médios- Transformando em Array ou Lista
x = raw_data['Revenue'].values
sns.distplot(x, color='blue')
#Adicionando uma linha na ertical com o valor médio
plt.axvline(media,0,1, color='red')
```

```python
#3- Vamos plotar a distribuição dos Visitantes (Visitors) - Transformando em Array ou Lista
x = raw_data['Visitors'].values

#Calculando a média
media = raw_data['Visitors'].mean()
media

sns.distplot(x, color='red')
#Adicionando uma linha na ertical com o valor médio
plt.axvline(media,0,1, color='blue')
```
