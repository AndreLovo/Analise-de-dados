Importando as bibliotecas

```
import numpy as np
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt

```

Histograma com Seaborn

```
x = np.random.normal (size=100) #Gerando 100 números aleatórios na distribuição nornmal
```

```
x
```

```
#Gerando o histograma - Agrupamento de faixas de valores - distribuição de valores
#kde= Linha traçada
#rug= Mostra os registros na linha horizontal
#hist=Gera o histograma
#bins=Mostra a quantidade de barrinhas verticais

sns.distplot(x, kde= True, rug=True, hist=True, bins=20)
plt.grid(axis='y', alpha=0.75)
plt.title('Histograma de valores Randômicos para trabalho com o Seaborn', fontsize=18, color='white')
plt.xticks (color='white')
plt.yticks (color='white')
plt.xlabel('valores da Idade',color='white')
plt.ylabel('valores de y',color='white')
plt.show()
```

```
# Gerando Gráfico de linha com 2 variáveis (utilizamos os eixos x e y)
df = pd.DataFrame(dict(time=np.arange(500),
                       value=np.random.randn(500).cumsum()))
g = sns.relplot(x="time", y="value", kind="line", data=df)
g.fig.autofmt_xdate()
plt.grid(axis='y', alpha=0.75)
plt.title('Gráfico de Linha para os valores Randômicos para trabalho com o Seaborn', fontsize=18, color='orange')
plt.xticks (color='orange')
plt.yticks (color='orange')
plt.xlabel('valores randômicos',color='orange')
plt.ylabel('valores de y',color='orange')
```
