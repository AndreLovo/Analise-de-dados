```
# Importando as dependencias do Seaborn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# Importando os dados do google Colab

df_flight = pd.read_csv('/content/drive/My Drive/Colab Notebooks/seaborn-data-master/flights.csv')
df_flight
```

```
# Função relplot()
# Método de alto nível para visualização de relações estatísticas entre variáveis. 
# O método relplot() já combina o scatter plot e line plots. 
```

```
#Imprimindo o gráfico 1
sns.relplot(x="passengers", y="month", data=df_flight)
plt.xticks (color='white')
plt.yticks (color='white')
plt.ylabel('month',color='orange')
plt.xlabel('passengers',color='orange')
plt.show()
```

```
#Imprimindo o gráfico 2
sns.relplot(x="passengers", y="month", hue="year", data=df_flight)
plt.xticks (color='white')
plt.yticks (color='white')
plt.ylabel('month',color='orange')
plt.xlabel('passengers',color='orange')
plt.show()
```

```
# Importando os dados do google Colab

df_tips = pd.read_csv('/content/drive/My Drive/Colab Notebooks/seaborn-data-master/tips.csv')
df_tips
```

```
# Imprimindo o gráfico scatter 1 - relplot
sns.relplot(x="total_bill", y="tip", data=df_tips)
plt.show()
```

```
# Visualizando a relação entre as variáveis total da conta e gorjeta e 
# colorindo cada ponto com a variável Sex. - relplot

sns.relplot(x="total_bill", y="tip", hue="sex", style="sex", data=df_tips);
plt.show()
```

```
#Imprimindo o grafico pelo tip x total_bill (size) - relplot
sns.relplot(x="total_bill", y="tip", hue="size", data=df_tips)
plt.show()
```

```
#Imprimindo o gráfico 3 - relplot
sns.relplot(x="time", y="tip", data=df_tips, kind="line")
plt.show()
```

```
# Imprimindo usando o gráfico catplot (mudando a função para violino)
sns.catplot(x="day", y="total_bill", kind="violin", data=df_tips)
plt.show()
```

```
# Imprimindo usando o gráfico catplot (mudando a função para violino quadrado ou caixa ou boxen)
sns.catplot(x="day", y="total_bill", kind="boxen", data=df_tips)
plt.show()
```

```
#Plotando outro grafico seaborn usando o boxplot
sns.set(style="white", color_codes=True)
sns.boxplot(x="day", y="total_bill", data=df_tips)
plt.show()
```

```
# Scatterplot - 1
# Caso não queira usar o relplot() o método scatterplot() já atende 
# perfeitamente quando queremos ver relações nos dados.
# A única diferença é que o scatterplot plota gráfico de dispersão 
#apenas, logo este NÃO contém a função lineplot(). 

sns.scatterplot(x="total_bill", y="tip", hue="day", style="time", data=df_tips)
plt.show()
```

```
# Redimensionando Figuras no Seaborn
# O seaborn herda as funções do matplotlib, permitindo que você 
# controle o tamanho da figura, o tamanho dos eixos entre outros.
import matplotlib.pyplot as plt
f, ax = plt.subplots(figsize=(12, 7))
sns.scatterplot(x="total_bill", y="tip", hue="day", data=df_tips)
plt.show()
```

```
# Gráfico usando a paleta de cor sequencial “Blues”

sns.relplot(x="total_bill", y="tip", hue="size", palette="Blues", size="size", sizes=(12,250), data=df_tips)
plt.show()
```

```
# Gráfico usando a paleta de cor qualitativa “Set2”.

sns.catplot(x="sex", kind="count", palette="Set2", data=df_tips)
plt.show()
```

```
# Gráfico usando a paleta de cor divergente “PiYG”.
# Mapa de calor
sns.heatmap(df_tips.corr(), annot=True, cmap="PiYG")
plt.show()
```

```
#Definindo cores padrão para gráficos Searborn.
sns.set_palette("pastel")
#plot usando a paleta de cor "pastel" - a cor fica mais clara
sns.relplot(x="total_bill", y="tip", data=df_tips);
plt.show()
```

```
#Instrução with para definir cor temporariamente.
# Plota o gráfico definindo uma opção de cor temporária
sns.color_palette("Accent")
sns.relplot(x="total_bill", y="tip", size="size", sizes=(12,250), data=df_tips)
plt.show()
```

```
# Plotando dados categóricos

# Quando trabalhamos com variáveis categóricas temos que visualizar dados de formas diferentes.
# O Seaborn fornece o método catplot() que já contém diversos tipos de gráficos embutidos.
# Isso facilita, pois, você pode usar diferentes gráficos usando um mesmo método. 

# Gráficos de barras ou também conhecidos como gráficos de colunas.

sns.catplot(x="sex", kind="count", palette="Set2", data=df_tips)
plt.show()
```

```
# Gráfico de barras horizontais usando a coluna day

sns.catplot(y="day", kind="count", palette="Set1", data=df_tips)
plt.show()
```

```
# Scatter plot com dados categóricos
sns.catplot(x="day", y="total_bill", palette="Set2", data=df_tips)
plt.show()
```

```
# O parâmetro swarm evita sobreposição de pontos.

sns.catplot(x="day", y="total_bill", kind='swarm', data=df_tips)
plt.show()
```

```
# O parâmetro hue permite adicionarmos uma terceira variável a nossa visualização.
sns.catplot(x="day", y="total_bill", kind='swarm', hue="sex", palette="Dark2", data=df_tips);
plt.show()
```

```
# O parâmetro order permite alterarmos a ordem padrão das categorias que estão sendo exibidas. 
sns.catplot(x="day", y="total_bill", kind='swarm', hue="sex", palette="Dark2", order=['Sat','Sun','Fri','Thur'], data=df_tips);
plt.show()
```

```
# Invertendo a visualização para plots horizontais.
sns.catplot(y="day", x="total_bill", kind='swarm', hue="sex", palette="Dark2", order=['Thur','Fri','Sat','Sun'], data=df_tips);
plt.show()
```

```
# Gráficos com Regressão

# Quando temos muitas variáveis quantitativas em nossos dados é interessante visualizar como estas se relacionam.
# Podemos visualizar essas informações com linhas de regressão.
# Com modelos de regressão simples é possível checar se existe alguma correlação entre algumas variáveis. 
# Exibe a linha de regressão para visualizar a dispersão entre as variáveis.

%matplotlib inline

sns.lmplot(x="total_bill", y="tip", hue="sex", palette="Pastel1", data=df_tips)
plt.show()
```

```
# Usando o parâmetro col para segregar os gráficos pelo valor da coluna categórica time.

%matplotlib inline
sns.lmplot(x="total_bill",y="tip", hue="smoker", col="time", palette="Set2", data=df_tips)
plt.show()
```

```
# Na variável time temos 2 valores distintos Lunch e Dinner, conseguimos 
# segmentar os gráficos de correlação por coluna.

# Parâmetros scatter_kws e line_kws.

%matplotlib inline
sns.lmplot(x="total_bill",
                  y="tip", 
                 hue="smoker", 
                 col="time", 
                 palette="Set1",
                 data=df_tips, 
                 scatter_kws={'s': 90, 'alpha': 0.5},
                 line_kws={'lw': 4})
plt.show()
```

```
# Segregando gráficos pelo parâmetro col e pelo parâmetro row .

%matplotlib inline
sns.lmplot(x="total_bill", y="tip", hue="smoker",col="time", row="sex", palette="Set2", data=df_tips)
plt.show()

# Agora a segmentação é por horário de refeição e pelo sexo, 
# col faz a segmentação por coluna e row a segmentação por linha.
```
