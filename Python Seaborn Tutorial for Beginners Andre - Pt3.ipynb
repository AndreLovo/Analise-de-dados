## 1- Sub Plots

```python
# Usamos o Sub Plots quando queremos plotar diferentes tipos de Gráficos na mesma área
#1º Temos que setar a figura do Matplotlib figure
#2º Separando quais tipos de gráficos vamos colocar (Entender o que queremos mostrar)
#3ª Separa as variáveis numéricas (Visitors - Revenue - Marketing Spend) / (Clientes - Receitas - Gastos com Marketing)
```

```python
#Montando a área e a quantidade de gráficos a serem mostrados
fig, axis = plt.subplots(2,2, figsize =(12,7))
a = raw_data['Revenue'].values
b = raw_data['Visitors'].values
c = raw_data['Marketing Spend'].values

# Gráfico1 - Histograma das Receitas
sns.distplot(a, color='blue', ax = axis[0,0])

#Gráfico2 - Histograma dos Clientes
sns.distplot(b, color='green', ax = axis[0,1])

#Gráfico3 - Histograma dos gastos com Marketing
sns.distplot(c, color='red', ax = axis[1,0])

#Gráfico4 - Box Plot - 
# Vamos investigar as Receitas x dias x pontos x cores
# Investigação da distribuição dos pontos dos dados
ax = sns.boxplot(x='Revenue', y='Day_Name', data = raw_data, color= '#FFFF00')
ax = sns.swarmplot(x='Revenue', y='Day_Name', data = raw_data, color= 'red')
```

## 2- Pair Plots

```python
# Uma das técnicas mais importantes para a apresentação e análise dos resultados
```

```python
#1- Geral
# Vamos colocar no gráfico a planilha toda (Todas as colunas) - Vamos imprimir a planilha toda
# Distplot no meio (Azul)
# Gráfico de Distribuição em verde
g=sns.pairplot(raw_data, plot_kws={'color':'green'})
# Precisamos 'quebrar' as colunas e plotar somente as variáveis que são importantes e que não tenham redundância
# Plotar da forma abaixo não é comum
```

```python
#2-Vamos montar os gráficos com as colunas que tenham valores numéricos (Visitors - Revenue - Marketing Spend) / (Clientes - Receitas - Gastos com Marketing)
# vamos quebrar as colunas da planilha
# A intenção é investigar os dados usando os gráficos de Dispersão (scatter plot) e Distribuição (distribuition) POR LINHA
g=sns.pairplot(raw_data[['Visitors','Revenue','Marketing Spend']], plot_kws={'color':'green'}, height = 7)
```

```python
#3-Vamos agora plotar os gráficos usando o hue (legendas) de forma a facilitar nossas análises
g=sns.pairplot(raw_data[['Visitors','Revenue','Marketing Spend','Promo']], height = 5, hue='Promo')
# Com somente uma linha de programação montamos gráficos de acordo com as categorias de promoções(sem promoção - promoção red - promoção blue)
# Conseguimos investigar os relacionamentos muito rápido usando as colunas numéricas
```

```python
#4-Vamos agora plotar os gráficos de regressão - usando o hue (legendas) de forma a facilitar nossas análises daas colunas numéricas
g=sns.pairplot(raw_data[['Visitors','Revenue','Marketing Spend','Promo']], height = 5, hue='Promo', kind='reg')
```

## 3- Joint Plots - Gráficos Lineares (Juntar variáveis ou colunas da tabela)

```python
# Técnica usada para investigarmos o relacionamento entre duas variáveis com gráficos com duas variáveis (bivariável) e/ou uma variável
```

```python
#1- Vamos ver o relacionamento das Receitas com os investimentos em Marketing
ax = sns.jointplot('Marketing Spend','Revenue',data = raw_data, color = 'green',height = 10)
# Na linha horizontal de cima do gráfico temos a coluna Marketing Spend (Gastos com Marketing)
# Na linha vertical da lateral do gráfico temos a coluna Revenue (Receitas)
```

```python
#2- Vamos ver o relacionamento das Receitas com os investimentos em Marketing - vamos adicionar o gráfico de distribuição
ax = sns.jointplot('Marketing Spend','Revenue',data = raw_data, color = 'green', height = 10, kind = 'reg', ci=None)
```

## 4- Heat Map (Gráfico do Mapa de Calor) 

```python
# 1º Devemos criar um dataset para mostrarmos o mapa de calor - Usaremos o dataset de correlação
# .corr() é usado para encontrarmos correlações em par (duplas) - aos pares - usaremos as colunas da planilha.(Vamos usar as colunas numéricas) 
# Todos os valores nulos devem ser (são descartados) descartados.
# Quanto mais perto do valor 1, melhor serão os nossos dados.
```

```python
# Calculando as correlações entre as 3 colunas (Clientes - Receitas - Gastos com Marketing) - Valores entre -1 e 1
# estas análise vão depender das qualidades dos dados
pc = raw_data[['Visitors','Revenue','Marketing Spend']].corr()
pc
```

### Pesquisar sobre as correlações:

<https://statisticsbyjim.com/basics/correlations/>

```python
#hm = Heat Map
col = ['Visitors_PC','Revenue_PC','Marketing Spend PC']

hm = sns.heatmap(pc,
                annot = True,
                annot_kws = {'size':25},
                xticklabels = col,
                yticklabels = col 
                )

# annot = anotação no gráfico (colocar o valor no gráfico)
# Quanto mais escuro, menor a correlação (valor inicial=0.573842)
# Quanto mais claro, maior a correlação
# Podemos alterar o nome dos eixos - criando colunas
```

```python
# 2- Montando o Gráfico de Calor, alterando o cmap
hm = sns.heatmap(pc,
                annot = True,
                annot_kws = {'size':25},
                xticklabels = col,
                yticklabels = col,
                cmap = 'Greens')
```

```python
# 2- Montando o Gráfico de Calor, alterando o cmap
hm = sns.heatmap(pc,
                annot = True,
                annot_kws = {'size':25},
                xticklabels = col,
                yticklabels = col,
                cmap = 'Blues')
```

```python
# 2- Montando o Gráfico de Calor, alterando o cmap
hm = sns.heatmap(pc,
                annot = True,
                annot_kws = {'size':25},
                xticklabels = col,
                yticklabels = col,
                cmap="YlGnBu")
```
