## 5- LM Plots (é a combinação dos gráfico de regressão e grid Plots) - Scatter plots + Grid

```python
#1- Vamos investigar os gastos com Marketing x Receitas
ax = sns.lmplot(x = 'Marketing Spend', y = 'Revenue', data = raw_data, height = 10)
# Regressão linear é um método estatístico preditivo para modelar relacionamentos entre x (Gastos com marketing) e y (Receitas)
# Vamos tentar prever uma relação (Quanto maior o gasto com Marketing maior a Receita) 
# Podemos pesquisar mais em: 
# https://towardsdatascience.com/linear-regression-using-python-b136c91bf0a2 
```

```python
#2- Vamos fazer o mesmo Gráfico Investigando o investimento em Marketing x Receitas, alterando as cores, as legendas e o estilo
ax = sns.lmplot(x = 'Marketing Spend', y = 'Revenue', data = raw_data, height = 10, hue = 'Promo')
```

```python
#3- Vamos fazer o mesmo Gráfico Investigando o investimento em Marketing x Receitas, alterando as cores, as legendas e o estilo
# Vamos retirar os intervalos de segurança (ci = confidence interval)
ax = sns.lmplot(x = 'Marketing Spend', y = 'Revenue', data = raw_data, height = 10, hue = 'Promo', ci=None, markers = ["o", "x", "+"])
```

```python
#3- Vamos agora "quebrar" as análises em 3 diferentes gráficos: Investimento (Gastos) Marketing x tipo de promoção
# Vamos retirar os intervalos de segurança (ci = confidence interval)
# Vamos ver se melhora a visualização dos dados
ax = sns.lmplot(x = 'Marketing Spend', y = 'Revenue', data = raw_data, height = 5, col = 'Promo', ci=None, 
                line_kws = {'color':'red'},
                scatter_kws = {'color':'#00FF00'}
               )
```

```python
#4- Vamos agora "quebrar" as análises em 7 diferentes gráficos (cada dia da semana): Investimento (Gastos) Marketing x Dia da semana
# Vamos retirar os intervalos de segurança (ci = confidence interval)
# Vamos ver se melhora a visualização dos dados
ax = sns.lmplot(x = 'Marketing Spend', y = 'Revenue', data = raw_data, height = 5, col = 'Day_Name', ci=None, 
                line_kws = {'color':'blue'},
                scatter_kws = {'color':'#FFA07A'},
                #Vamos quebrar as colunas em 3
                col_wrap=3)
```
