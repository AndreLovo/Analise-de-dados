```python
#### teste
geolocator = Nominatim(user_agent="myApp")
from geopy.extra.rate_limiter import RateLimiter
geolocator = RateLimiter(geolocator.geocode, min_delay_seconds=1)

#### Aproximadamente 23 min.
```

```python
# Criando duas colunas latitude e longitude - inicilamente vazias
bairros['latitude']= ""
bairros['longitude']= ""

bairros.head()
```

```python
geolocator = Nominatim(user_agent="myApp")

for i in bairros.index:
    try:
        location=geolocator.geocode(bairros['CEP'][i])
        
        bairros.loc[i,'latitude']= location.latitude
        bairros.loc[i,'longitude']= location.longitude
    except:
        bairros.loc[i,'latitude']= ""
        bairros.loc[i,'longitude']= ""
        
bairros.head()
```

```python
# Fazendo a conversao
bairros['latitude'] = bairros['latitude'].astype(float)
bairros['longitude'] = bairros['longitude'].astype(float)
```
