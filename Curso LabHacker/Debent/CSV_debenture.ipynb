```python
#Vamos importar apemas algumas colunas do arquivo
#0-data, #1-emissor, #4-qtd, #5-número de negociações, #10-setor

#carregar o arquivo csv mas somente as colunas
df = pd.read_csv('Debenture.csv', sep=';', usecols=[0, 1, 2, 4, 5, 10])

#Trocando o nome das colunas
df.columns = ['Datas', 'Emissores', 'Ativos', 'Quantidades', 'Negociações','Setores']

#Exibir as primeiras linhas do DataFrame
df.head(5)
```

```python
#Descrevendo as colunas numéricas
df.describe()
```

```python
#Contando os ativos de cada emissor
df['Ativos'].value_counts()
```

```python
#Plotar o gráfico por Sub Grupo de Ativos - Grafico de barras
df['Ativos'].value_counts().plot.bar(title='Quantidades de Debentures por Ativos')
```

```python
#Agora vamos pegar os setores e mostrar em um gráfico de barras
df['Setores'].value_counts().plot.bar(title='Quantidades de Debentures por Setor')
```

```python
#Mostrar todos os dados das maiores negociações por setores
df.loc[df.groupby('Setores')['Quantidades'].idxmax()]
```

```python
#Selecionando uma faixa (de 68000 até 1500000) de Quantidades de debentures 
df[(df['Quantidades'] >68000) & (df['Quantidades']<1500000)]
```
